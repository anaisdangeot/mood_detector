```python
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud,ImageColorGenerator
import sys,os
import numpy as np
import pandas as pd
from PIL import Image 
```

```python
text = open('/home/matt/code/anaisdangeot/mood_detector/code/api/lyrics_parts_test.txt',mode="r",encoding='utf-8').read()
python_mask = np.array(Image.open("/home/matt/code/anaisdangeot/mood_detector/code/api/smile_318-348868.png").convert("RGBA"))
python_mask[np.all(python_mask == [255, 255, 255, 255], axis=2)] = [0, 0, 0, 0]
plt.figure(figsize=(10,10),facecolor='none')
WordCloud = WordCloud(background_color='rgba(0,0,0,0)', mask=python_mask,height=512,width=512).generate(text)
plt.axis('off')
plt.imshow(WordCloud,interpolation='bilinear')
```

```python
text = open('/home/matt/code/anaisdangeot/mood_detector/code/api/lyrics_parts_test.txt', mode="r", encoding='utf-8').read()
python_mask2 = np.array(Image.open("/home/matt/code/anaisdangeot/mood_detector/code/api/sad-emoticon-face_318-54034.png").convert("RGBA"))
python_mask2 = np.array(Image.fromarray(python_mask2).resize((650, 650)))  # Resize the image mask

plt.figure(figsize=(10, 10), facecolor='none')
wordcloud = WordCloud(background_color='rgba(0,0,0,0)', mask=python_mask2, height=650, width=650).generate(text)

plt.axis('off')
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```

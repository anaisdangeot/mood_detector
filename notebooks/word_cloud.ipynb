```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud,ImageColorGenerator
import sys,os
import numpy as np
import pandas as pd
from PIL import Image 
```

```python
text = open('/home/matt/code/anaisdangeot/mood_detector/code/api/lyrics_parts_test.txt',mode="r",encoding='utf-8').read()
positive_mask = np.array(Image.open("/home/matt/code/anaisdangeot/mood_detector/code/api/smile_318-348868.png").convert("RGBA"))
positive_mask[np.all(positive_mask == [255, 255, 255, 255], axis=2)] = [0, 0, 0, 0]
plt.figure(figsize=(10,10),facecolor='none')
WordCloud = WordCloud(background_color='rgba(0,0,0,0)', mask=positive_mask,height=1000,width=1000).generate(text)
plt.axis('off')
plt.imshow(WordCloud,interpolation='bilinear')
```

```python
text = open('/home/matt/code/anaisdangeot/mood_detector/code/api/lyrics_parts_test.txt', mode="r", encoding='utf-8').read()
negative_mask = np.array(Image.open("/home/matt/code/anaisdangeot/mood_detector/code/api/sad-emoticon-face_318-54034.png").convert("RGBA"))
negative_mask= np.array(Image.fromarray(negative_mask).resize((650, 650))) 

plt.figure(figsize=(10, 10), facecolor='none')
wordcloud = WordCloud(background_color='rgba(0,0,0,0)', mask=negative_mask, height=650, width=650).generate(text)

plt.axis('off')
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```
